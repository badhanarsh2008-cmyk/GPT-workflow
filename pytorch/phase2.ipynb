--------Backward----------

like who is responsible for loss and how much

```python
a = torch.tensor(3.0 ,requires_grad=True)
w = torch.tensor(2.0)
y_pred = a * w 
y_true = torch.tensor(10.0)

loss = (y_pred - y_true ) ** 2
print(loss)
loss.backward()
print(a.grad)

#means yaha prr a ki wajeh se glt ans aya so its telling us a = -16 the reason behind wrong guess.
```

--------.Grad------

how much i should change to reduce the change.
like backward ke baad AI ko pta chl jata hai ki woh glt hai but usko btana pdta hai ki kis direction(increase/decrease) mein glt hai.

w.grad > 0          then it decreases the value of w.
w.grad < 0          then it increases the value of w.

```python
a = torch.tensor(3.0 ,requires_grad=True)
w = torch.tensor(2.0)
y_pred = a* w 
y_true = torch.tensor(10.0)

loss = (y_pred - y_true ) ** 2
print(loss)
loss.backward()
print(a.grad)

print("Prediction :",y_pred.item())
print("Loss :",loss.item())
print("w.grad : ",a.grad.item()) #it gives negative so means we have to increase the value of it 
```

Gradient

Gradient descent


like number ko change krna according to advice ... ki number ko kitne step increase krna hai ya decrease so woh accurate value ke paas jaaye

old_val = current guess
new val = output or ans after eqn
gradient = .grad
learning_rate = lr

```python
#new_val = old_val - (learn_rate * gradient)


a = torch.tensor(2.0)
w = torch.tensor(3.0 , requires_grad=True)

y_pred = a*w
y_true = torch.tensor(10.0)

#lr rate
lr = 0.1

loss = (y_pred - y_true) **2
loss.backward()

print("Before update.")
print("predicted value :",y_pred.item()) #we use item bcz in torch they all arent just float values they are actly objects so .item() returns us simple number value
print("loss :",loss.item())
print("gradient :",w.grad.item())

#update weight
with torch.no_grad():
    w-= y_pred - (lr - w.grad)
#important: gradient reset
w.grad.zero_() # ye overlap kr deta hai purani values ko dlt kr deta hai and nye sire se shuruat krta hai 
print("After update!")
print("w :",w.item())

```

-----------------------------TRAINING LOOP---------------------------------

1.make prediction
2.measuring loss
3.find
4.fix
5.clear

```python
#dataset
x = torch.tensor([1.,2.,3.,4.])
y_true = torch.tensor([2.,4.,6.,8.])

w = torch.tensor(0.0,requires_grad=True)
lr = 0.1
epochs = 10

for epoch in range(epochs):
    #prediction
    y_pred = w * x
    #loss count
    loss = ((y_pred - y_true) ** 2).mean()
    loss.backward()
    
    with torch.no_grad():
        w -= lr * w .grad
    
    print(f"Epoch {epoch + 1},w = {w.item():.4f},loss = {loss.item():.4f}")

    
```
