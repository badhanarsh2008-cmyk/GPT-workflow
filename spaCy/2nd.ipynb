STOPWORDS : These words are frequently used in a language but typically do not contribute much to the meaning of a sentence.  In natural langugae processing ( NLP ), stopwords are often removed from text data during the pre processing to focus on more meaningful words and improve the efficiency of analysis

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
text = """
Apple was founded by Steve Jobs in California in 1976. 
The company is now worth over 3 trillion dollars.
"""

doc = nlp(text)

print("=== ENTITIES FOUND ===")
for ent in doc.ents:
    print(f"{ent.text:20} → {ent.label_}")
```

```python
print("\n=== KEYWORDS ===")
keywords = [token.text for token in doc 
            if token.pos_ in ["NOUN", "PROPN"] 
            and not token.is_stop]
print(keywords)
```

```python
print("\n=== SENTENCES ===")
for sent in doc.sents:
    print("-", sent.text.strip())
```
