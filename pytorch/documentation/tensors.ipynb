PYTORCH [ Tensor ]

Tensor : Tensors are specialized data structure that are very similar to arrays and matrices. In pytorch we use tensors to encode the input and output of a model, as well as the models parameters.

its like numpy ndarrays. except that the tensors can run at gpu and other hardware accelerators. fact, tensors and NumPy arrays can often share the same underlying memory, eliminating the need to copy data (see Bridge with NumPy). Tensors are also optimized for automatic differentiation (we’ll see more about that later in the Autograd section). If you’re familiar with ndarrays, you’ll be right at home with the Tensor API.
