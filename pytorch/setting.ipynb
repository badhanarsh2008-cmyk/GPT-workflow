Linear Regression

```python
x = torch.tensor([1.0,2.0,3.0])
y = torch.tensor([4.0,5.0,6.0])

#wrong guessing 
w = torch.tensor(0.0)
s = torch.tensor(0.0)

#step-4
y_pred = w * x + s

loss = ((y_pred - y) ** 2).mean()   #used square for positive outcome and mean to arrange bigger values 

print("Predicted Values :",y_pred)
print("Actual Values :",y)
print("Losses:",loss)
```
